### Pitchfork

```python
pitchfork = requests.get('https://pitchfork.com/features/lists-and-guides/7693-the-top-500-tracks-of-the-2000s-20-1/')
pitchfork.status_code
```

```python
fork = BeautifulSoup(pitchfork.content, 'html.parser')
```

```python
fork_list = fork.find_all('div', attrs={'class': "blocks-area"})
fork_list
```

```python
def get_songs(artists, titles):
    artists = [a.get_text() for x in fork_list for a in x.find_all('ul', attrs={'class': 'artist-list'})]
    titles = [a.get_text() for x in fork_list for a in x.find_all('h2', attrs={'class': 'list-blurb__work-title'})]
    
    dct = {'song_title': titles, 'artist': artists}
    
    return dct
    
```

```python
starts = range(1, 12)
list(starts)
```

```python
PitchforkTop200 = pd.DataFrame()

for start in starts:
    pitchfork = requests.get(f'https://pitchfork.com/features/lists-and-guides/7693-the-top-500-tracks-of-the-2000s-20-1/?page={start}')
    fork = BeautifulSoup(pitchfork.content, 'html.parser')
    fork_list = fork.find_all('div', attrs={'class': "blocks-area"})
    info_dct = get_songs(artists, titles)
    new_df = pd.DataFrame.from_dict(info_dct)
    PitchforkTop200 = pd.concat([PitchforkTop200, new_df])



PitchforkTop200




```
